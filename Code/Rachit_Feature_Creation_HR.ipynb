```python
import numpy as np
import pandas as pd
```

```python
X_train = pd.read_csv('./UltimateFinal/ratings_train.csv')
```

```python
X_test = pd.read_csv('./UltimateFinal/ratings_test.csv')
```

```python
df_books = pd.read_csv('./UltimateFinal/Books.csv')
```

```python
df_users = pd.read_csv('./UltimateFinal/Users.csv')
```

```python
df_users.fillna(0, inplace=True)
```

```python
df_books_onehot = pd.get_dummies(df_books)
```

```python
df_users.columns = ['User', 'Arts & Photography', 'Biographies & Memoirs',
       'Business & Money', 'Childrens Books', 'Christian Books & Bibles',
       'Comics & Graphic Novels', 'Computers & Technology', 'Cookbooks',
       'Crafts', 'Education & Teaching', 'Engineering & Transportation',
       'Health', 'History', 'Humor & Entertainment', 'Law', 'Lesbian',
       'Literature & Fiction', 'Medical Books', 'Mystery',
       'Parenting & Relationships', 'Politics & Social Sciences', 'Reference',
       'Religion & Spirituality', 'Romance', 'Science & Math',
       'Science Fiction & Fantasy', 'SelfHelp', 'Sports & Outdoors',
       'Teen & Young Adult', 'Travel', 'Avg Price']
```

```python
df_users
```

```python
X_train_ufeatures = X_train.merge(df_users, how='left', on='User')
X_test_ufeatures = X_test.merge(df_users, how='left', on='User')
```

```python
# X_train_ufeatures = X_train.merge(df_users, how='left', on='User')
X_train_ui_features = X_train_ufeatures.merge(df_books_onehot, how='left', on='Item')
X_test_ui_features = X_test_ufeatures.merge(df_books_onehot, how='left', on='Item')
# X_test_ufeatures = X_test.merge(df_users, how='left', on='User')
```

```python
X_train_ui_features.to_csv('./UltimateFinal/ratings_train_with_user_item_features.csv')
X_test_ui_features.to_csv('./UltimateFinal/ratings_test_with_user_item_features.csv')
```

```python
y_train = X_train_ui_features['Rating']
y_test = X_test_ui_features['Rating']
```

```python
X_train = X_train_ui_features.drop('Rating', axis=1)
X_test = X_test_ui_features.drop('Rating', axis=1)
```

```python
X_train.to_csv('./UltimateFinal/Regression/X_train.csv')
y_train.to_csv('./UltimateFinal/Regression/y_train.csv')
X_test.to_csv('./UltimateFinal/Regression/X_test.csv')
y_test.to_csv('./UltimateFinal/Regression/y_test.csv')
```
