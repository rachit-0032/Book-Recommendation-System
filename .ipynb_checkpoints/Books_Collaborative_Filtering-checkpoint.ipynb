```python
temp = ratings.groupby('User-ID').count()['Book-Rating']>200
pop_users = temp[temp].index
```

```python
filtered_rating = ratings[ratings['User-ID'].isin(pop_users)]
```

```python
temp = filtered_rating.groupby('ISBN').count()['Book-Rating']>50
famous_books = temp[temp].index
```

```python
ratings_final = filtered_rating[filtered_rating['ISBN'].isin(famous_books)]
```

```python
len(ratings_final)
```

```python
ratings_matrix = ratings_final.pivot_table(index='User-ID',columns='ISBN',values='Book-Rating')
ratings_matrix.fillna(0,inplace=True)
```

```python
ratings_matrix.shape
```

```python
from scipy import sparse
```

```python
sparse_ratings_matrix = sparse.csr_matrix(ratings_matrix)
```

```python
global_average_rating = sparse_ratings_matrix.sum()/sparse_ratings_matrix.count_nonzero()
```

```python
global_average_rating
```

```python
def get_average_rating(sparse_matrix, is_user):
    ax = 1 if is_user else 0
    sum_of_ratings = sparse_matrix.sum(axis = ax).A1  
    no_of_ratings = (sparse_matrix != 0).sum(axis = ax).A1 
    rows, cols = sparse_matrix.shape
    average_ratings = {i: sum_of_ratings[i]/no_of_ratings[i] for i in range(rows if is_user else cols) if no_of_ratings[i] != 0}
    return average_ratings
```

```python
average_rating_user = get_average_rating(sparse_ratings_matrix, True)
```

```python
avg_rating_movie = get_average_rating(sparse_ratings_matrix, False)
```

```python
U, sigma, Vt = svds(ratings_matrix, k = 100)
```

```python
def cosine_similarity(v,u):
    return (v @ u)/ (np.linalg.norm(v) * np.linalg.norm(u))
```

```python
# highest_similarity = -np.inf
# highest_sim_book = -1
book = 3
similarities = {}
for b in range(0,Vt.shape[1]):
    similarity = cosine_similarity(Vt[:,book], Vt[:,b])
    
    similarities[b]=similarity
    
#     if similarity > highest_similarity:
#         highest_similarity = similarity
#         highest_sim_col = col
```
